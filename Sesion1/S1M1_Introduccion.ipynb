# **2. NOCIONES GENERALES DE PROGRAMACIÓN**

## **2.1. ¿Qué es programar?**

Programar consiste en producir un conjunto de instrucciones para realizar una tarea utilizando un lenguaje que puede entender un computador. 

Las intrucciones que pueden realizarse en un computador moderno son muy variadas:
* Operaciones matemáticas y lógicas sobre números 
* Procesar cadenas de texto u otras formas de organizar información más complejas
* Utilizar información almacenada en archivos 
* Controlar dispositivos de entrada para el registro de eventos físicos 
* Controlar dispositivos de salida para la presentación de información visual, sonora o táctil 
* Comunicarse con otras computadoras a través de la red 


En otras palabras, programar consiste en implementar un determinado algoritmo utilizando un lenguaje de programación concreto

### **Definiciones**

**Algoritmo**: secuencia no ambigua y finita de instrucciones que han de realizarse para resolver un determinado problema (ejemplo: una receta de cocina). Puede entenderse de manera general como un flujo controlado por intrucciones lógicas: 

### **Definiciones**

**Programa**: conjunto de algoritmos y estructuras de datos implementados en un lenguaje y materializados en el código fuente (source code). 

**Lenguaje de programación**: Lenguaje formal (es decir, un lenguaje con reglas sintácticas bien definidas) que permite codificar una serie de instrucciones (algoritmos) con el fin de controlar el comportamiento físico o lógico de un computador. 

Según el grado de proximidad del lenguaje formal utilizado con el lenguaje humano (normalmente el inglés), hablamos de lenguajes de que puede "comprender" un computador bien de manera directa en el caso de los lenguajes de bajo nivel (código máquina) o a través de programas intermedios que traducen intrucciones en lenguajes de alto nivel (más próximos al lenguaje natural) 

**Código máquina**: el código directamente interpretable por un circuito programable, controla directamente el hardware aprovechando el conocimiento exhaustivo de su constitución física y lógica.

**Compilador**: programa que traduce el código fuente de otro programa escrito en un lenguaje de programación de alto nivel a código máquina, que puede ser eje. El proceso de traducción de código fuente en código máquina se conoce como compilación.

**Intérprete**: programa informático capaz de analizar y ejecutar el código fuente de otro programa y traducirlo a código máquina a medida que sea necesario, típicamente, instrucción por instrucción, y normalmente no guardan el resultado de dicha traducción. 

**Estructura de Datos**: estructura abstracta diseñada para la organización, gestión y almacenamiento eficientes de la información en sistemas informáticos. Implica una colección de datos, las relaciones entre ellos y las funciones u operaciones que pueden aplicarse a los datos. Distintas estructuras de datos son más apropiados para distintas tareas, y el uso de una estructura frente a otras siempre conlleva un compromiso entre ventajas e inconvenientes.

## **2.2. Principios generales de diseño de software**

La ingeniería del software es una amplia y compleja area de conocimiento que se encarga de determinar las mejores formas de programar para producir software estable y eficiente, con el menor número de errores posibles y con el mínimo impacto sobre el resultado, que sea sencillo y barato de mantener en el tiempo y de reparar cuando se detecten errores. 

Queda muy lejos de los objetivos de este curso tratar una temática tan compleja. Sin embargo, a continuación planteo algunos principios generales que permiten orientar la labor del programador para simplificar la tarea.

**YAGNI - No lo vas a necesitar**

    Lo primero antes de emprender cualquier acción es preguntarse: ¿necesitas realmente lo que pretendes hacer? ¿Puedes NO hacerlo? ¿Estás perdiendo el tiempo construyendo algo que no necesitas?
    Si resulta que en efecto lo necesitas, ¿puedes resolverlo de otra forma más sencilla?
    ¿Puede reutilizar el trabajo preexistente; desde dentro de su código base o tal vez librerías externas?
    No reinventes la rueda. No la necesitarás.

**Piensa antes de codificar**

    Este es el principio de todos los principios, el patrón de todos los patrones. Pensar ininterrumpidamente en un problema es difícil, pero es esencial. Date un paseo si lo necesitas, para despejarte de distracciones y centrarte en el problema que tienes entre manos y en las posibles soluciones. Elabora un diseño completo antes de proceder a la implementación.
    Busca en Google soluciones ya existentes. A continuación, busca más en Google para encontrar una solución más optimizada. Por último, busca más en Google para comparar esas soluciones y decídete por la mejor.
    Escribe pseudocódigo. Te ayuda a pensar mejor, y es más fácil compartirlo con tus colegas o en línea (Reddit, StackOverflow, etc.), para pedir opiniones.

**KISS - Keep It Simple Stupid (Mantenlo simple, estúpido)**

    Favorece la simplicidad en tu diseño y código. Como dice el refrán: "Si no puedes explicarlo de forma sencilla, no lo entiendes lo suficientemente bien".
    El beneficio de introducir un patrón de diseño debe superar la complejidad que añade.
    No introduzcas patrones de código porque sí. Empieza a introducir patrones sólo cuando hagan las cosas más limpias y comprensibles.
    Lee y relee tu código para ver si puedes simplificarlo un poco en cada pasada.
    Consejo: Recuerda que los buenos libros no se escriben, se reescriben. Repasa tu código una y otra vez hasta que sea lo suficientemente sencillo y conciso. Yo a veces simplifico un párrafo entero de código en una o dos líneas, lo que añade una legibilidad masiva.

**DRY - No te repitas**

    Reutiliza la mayor parte posible de tu código. Para ello, es necesario modularizar el código:
        - Divide grandes trozos de código en funciones más pequeñas y reutilizables.
        - Agrupe las funciones relevantes en clases y archivos.
        - Agrupa las clases relevantes en espacios de nombres.
        - Publica todo este código relevante en forma de bibliotecas. Internas o de código abierto.
    Utiliza el código de otros siempre que puedas. Busca en Google bibliotecas de código abierto o con licencia.
    Consejo: "Antes de que el software pueda ser reutilizable, primero tiene que ser utilizable". - Ralph Johnson

**Principio de responsabilidad única**

    Cualquier función que escribas debe tener una sola responsabilidad. Por ejemplo, una función llamada "MostrarFecha", que imprime la fecha actual.
    Cualquier clase que escribas debe tener también un propósito singular. Por ejemplo, una clase "DatePicker" que encapsule los métodos necesarios para ayudar al usuario a elegir una fecha. Podría tener funciones para mostrar la fecha seleccionada, establecer un intervalo de fechas, etc.
    Lo mismo ocurre con las bibliotecas y los módulos. Deberían tener responsabilidad sobre una parte de la funcionalidad proporcionada por todo el software.
    Consejo: Cuando empiece a tener funciones relacionadas con la hora y la zona horaria, etc. en su clase "DatePicker", es una buena señal de que debería mover las funciones relacionadas con la hora a una clase "TimePicker".

managers**Separación de responsabilidades**

    Similar en espíritu al principio de responsabilidad única, pero en un espectro más amplio; es el principio de seccionar un programa en partes separadas de modo que cada parte aborde una preocupación diferente.
    Cada programa tiene múltiples preocupaciones: por ejemplo, obtener datos del servidor, validar esos datos y mostrar esos datos al usuario. Si seccionas tu aplicación de forma que tengas tres capas: una para la comunicación cliente-servidor, otra para la validación de datos y otra para mostrar la interfaz de usuario, conseguirás la separación de intereses en esa aplicación.
    Consejo: Debes tener cuidado de no filtrar el código relacionado con la interfaz de usuario a tus capas de controlador, modelo y validación. Además, ten cuidado de no filtrar la lógica del controlador al código de comunicación. Utilizar librerías bien pensadas como React para UI y gRPC para cliente-servidor puede ayudarte a reforzar estas separaciones.

## **4.3. Ipython: El intérprete interactivo _aumentado_**

Como habrás podido comprobar, el intérprete por defecto de python es demasiado simple para un trabajo interactivo cómodo. Para añadir muchas funciones y comodidades a dicho intérprete se recomienda utilizar ipython, un interprete de python _on steroids_. 

**Historial de comandos.** Al igual que un shell UNIX, la consola IPython soporta el historial de comandos. Utiliza las flechas arriba y abajo para navegar por los comandos escritos anteriormente.

**Completar con tabulador (tab completion)**. Es una forma cómoda de explorar la estructura de cualquier objeto con el que estés tratando. Simplemente escribe nombre_objeto.<TAB> para ver los atributos del objeto. Además de los objetos y palabras clave de Python, el tabulador también funciona con nombres de archivos y directorios.

**Funciones mágicas.** La consola y los notebooks admiten las llamadas funciones mágicas anteponiendo a un comando el carácter %. Entre la variedad de funciones mágicas disponibles en ipython, aquí tienes una lista de las más útiles:  

* **%cd**, para cambiar el directorio de trabajo.  
* **%paste**, pega un fragmento de código copiado anteriormente.
* **%cpaste**, permite pegar código, especialmente código de sitios web que ha sido prefijado con el prompt estándar de Python (por ejemplo >>>) o con un prompt de ipython, (por ejemplo en [3]).
* **%timeit**, para cronometrar la ejecución de fragmentos cortos utilizando el módulo timeit de la biblioteca estándar.
* **%debug**, entrar en el modo de depuración post-mortem. Es decir, si el código que intentas ejecutar, lanza una excepción, usando %debug entrarás en el depurador en el punto donde se lanzó la excepción.

**Alias.** Además IPython incluye varios alias que emulan las herramientas de línea de comandos UNIX más comunes, como _ls_ para listar archivos, _cp_ para copiar archivos y _rm_ para eliminar archivos (se muestra una lista completa de alias al escribir alias).

**Obtener ayuda.** Se puede acceder a la hoja de trucos incorporada a través de la función mágica **%quickref**. Una lista de todas las funciones mágicas disponibles se muestra al escribir **%magic**.

## **5. Generalidades acerca de la programación en python**

### **Indentación**
A diferencia de otros lenguajes que utilizan llaves para definir los bloques de código, cuando Guido Van Rossum creó el lenguaje quiso evitar estos caracteres por considerarlos innecesarios. Es por ello que en Python los bloques de código se definen a través de espacios en blanco, preferiblemente 4. Por ejemplo:
```
if edad > 3:
    print('La edad es mayor que tres')
else:
    print('La edad es igual o menor que tres')
```

También se pueden usar tabuladores para generar la indentación de los bloques de código (aunque no se recomienda). Es importante no mezclar en un mismo archivo espacios y tabuladores para indicar la indentación, porque es un error sintáctico. 

### **Ancho del código**

Los programas suelen ser más legibles cuando las líneas no son excesivamente largas. La longitud máxima de línea recomendada por la guía de estilo de Python es de 80 caracteres. Debido a la prevalencia de pantallas panorámicas, hay profesionales que recomiendan usar 100 caracteres de ancho de línea. En cualquier caso, no es un regla inviolable pero es importante evitar que las líneas sean demasiado anchas.

En la siguiente sección veremos métodos para dividir lineas largas en varias lineas.
